# Pareto

* Puede ser de interás para solucionar problemas de arranque en frio en ámbitos que un recomendador basado en contenidos no es viable.

* En este ejemplo aplicaremos Pareto para la generación de paquetes turisticos a recomendar

## Antes de comenzar... veamos el dataset

**Datos de medios de transporte**, en este caso para Jerez - Madrid

```python
import pandas as pd

transportes = pd.read_csv('./dataset/productosTransporteSTMate.csv', sep='|')
transportes.head(5)
```

**Datos de alojamientos**, en este caso para Madrid

```python
hoteles = pd.read_csv('./dataset/productosHotelSTMate.csv', sep='|')
hoteles.head(5)
```

## Comienzo de Pareto

### Medios de transporte

**Paso 1**. Ordenar por precios

```python
transportes = transportes.sort_values(['DblPrecio'])
transportes.head(5)
```

**Paso 2**. Eliminar usando la frontera inferior de Pareto

```python
import dateutil.parser
import datetime

#Función que calcula si un elemento esta dominado por otro
def dominaTransporte(T1, T2):
    DtmFechaHoraLlegadaVuelta = dateutil.parser.parse(T1['DtmFechaHoraLlegadaVuelta'])
    DtmFechaHoraSalidaIda = dateutil.parser.parse(T1['DtmFechaHoraSalidaIda'])
    
    duracionT1 = DtmFechaHoraLlegadaVuelta - DtmFechaHoraSalidaIda
    
    DtmFechaHoraLlegadaVuelta = dateutil.parser.parse(T2['DtmFechaHoraLlegadaVuelta'])
    DtmFechaHoraSalidaIda = dateutil.parser.parse(T2['DtmFechaHoraSalidaIda'])
    
    duracionT2 = DtmFechaHoraLlegadaVuelta - DtmFechaHoraSalidaIda
    
    dblUmbralTiempo = datetime.timedelta(minutes=30)
    
    #elemento 1 no es peor que elemento 2 en ningun objetivo
    blnCondicion1 = (not(T1['DblPrecio'] > T2['DblPrecio']) and not(abs(duracionT1 - duracionT2)<dblUmbralTiempo))
    #elemento 1 es mejor que elemento 2 en al menos un objetivo
    blnCondicion2 = (T1['DblPrecio'] < T2['DblPrecio'] or abs(duracionT1 - duracionT2) >= dblUmbralTiempo)
        
    return blnCondicion1 and blnCondicion2


lstNoDominados = list()
lstDominadosParaEliminar=list()
#for i in range(len(transportes)):
for i in range(30):
    blnDominado = False
    for ixND in range(len(lstNoDominados)):
        if(not dominaTransporte(transportes.iloc[ixND], transportes.iloc[i])):
            blnDominado = True
        if(dominaTransporte(transportes.iloc[i], transportes.iloc[ixND])):
            lstDominadosParaEliminar.append(transportes.iloc[ixND]);
    if(blnDominado == False):
        lstNoDominados.append(transportes.iloc[i])
    
    lstTmp = list()
    for iNoDominados in range (len(lstNoDominados)):
        iDominados = 0
        blnEncontrado = False
        
        while (iDominados < len(lstDominadosParaEliminar) and not blnEncontrado):
            if(lstNoDominados[iNoDominados]['_id'] == lstDominadosParaEliminar[iDominados]['_id']):
                blnEncontrado = True
            iDominados = iDominados+1
        if not blnEncontrado:
            lstTmp.append(lstNoDominados[iNoDominados])
    
    lstNoDominados = lstTmp;

lstTransporteFinal = lstNoDominados

print  (lstTransporteFinal)
```

## Alojamientos

**Paso 1**. Ordenar por precio

```python

hoteles = hoteles.sort_values(['DblPrecio'])
hoteles.head(5)
```

**Paso 2**. Eliminar usando la frontera inferior de Pareto

```python
#Función que calcula si un elemento esta dominado por otro
def dominaAlojamiento(T1, T2):
    #elemento 1 no es peor que elemento 2 en ningun objetivo
    blnCondicion1 = (not(T1['DblPrecio'] > T2['DblPrecio']) and not(T1['IntCategoria'] < T2['IntCategoria']));
    #elemento 1 es mejor que elemento 2 en al menos un objetivo
    blnCondicion2 = (T1['DblPrecio'] < T2['DblPrecio'] or T1['IntCategoria'] > T2['IntCategoria']);
    return blnCondicion1 and blnCondicion2;
    
lstNoDominados = list()
lstDominadosParaEliminar=list()
#for i in range(len(hoteles)):
for i in range(30):
    blnDominado = False
    for ixND in range(len(lstNoDominados)):
        if(not dominaAlojamiento(hoteles.iloc[ixND], hoteles.iloc[i])):
            blnDominado = True
        if(dominaAlojamiento(hoteles.iloc[i], hoteles.iloc[ixND])):
            lstDominadosParaEliminar.append(hoteles.iloc[ixND]);
    if(blnDominado == False):
        lstNoDominados.append(hoteles.iloc[i])
    
    lstTmp = list()
    for iNoDominados in range (len(lstNoDominados)):
        iDominados = 0
        blnEncontrado = False
        
        while (iDominados < len(lstDominadosParaEliminar) and not blnEncontrado):
            if(lstNoDominados[iNoDominados]['_id'] == lstDominadosParaEliminar[iDominados]['_id']):
                blnEncontrado = True
            iDominados = iDominados+1
        if not blnEncontrado:
            lstTmp.append(lstNoDominados[iNoDominados])
    
    lstNoDominados = lstTmp;

lstAlojamientoFinal = lstNoDominados

print  (lstAlojamientoFinal)
```
